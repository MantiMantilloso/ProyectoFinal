```python
import pandas as pd

# Query para obtener accidentes en Manhattan en 2024 con coordenadas válidas
url = ("https://data.cityofnewyork.us/resource/h9gi-nx95.json?"
       "$where=borough='MANHATTAN' AND latitude IS NOT NULL "
       "AND date_extract_y(crash_date) = 2024"
       "&$limit=5000")

df_crashes = pd.read_json(url)
```
