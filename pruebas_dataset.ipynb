```python
import pandas as pd
import requests

# Query para obtener accidentes en Manhattan en 2024 con coordenadas validas
base_url = "https://data.cityofnewyork.us/resource/h9gi-nx95.json"
params = {
    "$where": "borough='MANHATTAN' AND latitude IS NOT NULL AND date_extract_y(crash_date)=2024",
    "$limit": 5000,
}

response = requests.get(base_url, params=params, timeout=30)
response.raise_for_status()

df_crashes = pd.DataFrame(response.json())
df_crashes.head()
```
